```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Création d'un graphique
La commande de base est `plt.plot(x, y, formattage)`

x et y étant l'abscisse et l'ordonnée sous forme de np.array.

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-poster')
```

```python
lambda_ = 2
x = np.linspace(0, 5*lambda_, 256)
fig, axes = plt.subplots(3, 3, sharex='col', sharey='row')
phi= 0
A = 2
Pi = np.pi
for t in range(9):
    y = A*np.cos(2*Pi*x/lambda_ - phi)
    axes[t//3, t%3].plot(x, y)
    axes[t//3, t%3].scatter(t*lambda_/4, A, color='red')
    axes[t//3, t%3].scatter(t*lambda_/4+3*lambda_, A*math.cos(6*Pi), color='green')
    delta = 1.3
    axes[t//3, t%3].scatter(t*lambda_/4+delta*lambda_, A*math.cos(2*delta*Pi), color='black')
    phi += Pi/2 # à chaque 1/4 de période, le déphasage d'amplitude de chaque point du milieu est de Pi/2
    axes[t//3, 0].set(ylabel='Ampl')

axes[2, 1].set(xlabel='Distance (m)')  
axes[0, 1].set(title="Propagation d'une onde") 
plt.show()
```
