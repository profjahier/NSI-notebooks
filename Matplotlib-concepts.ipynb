# Concepts de base
Un "graphique" est strucutré de la manière suivante :
- `Figure` est l'objet de plus haut niveau : "coquille externe" du graphique
- `Axes` contient l'objet graphique en lui-même
- `Axis` est un axe du graphique (conrtôle des labels, xmin, xmax, etc.)
- `Artist` est chaque élément dessiné sur un graphique (ligne, titre, etc)
- `Canvas` est la zone de tracé du graphe

Les données à tracer doivent être de type `array` de `numpy`. (np.array)

Toute utilisation de matplotlib commence donc toujours par importer les modules `matplotlib.pyplot` et `numpy` sous la forme usuelle ci-dessous :

```python
import matplotlib.pyplot as plt
import numpy as np
```

Le module `pyplot` crée automatiquement (et de façon transparente) les objets `figure` et `axes` nécessaires à la construction d'un graphique. Chaque "élément" qu'on rajoute par la suite (ex : plt.plot() ou plt.title()) s'ajoute à la `figure`et l'`axe` actif.

Pour visualiser la figure créée, il faut utiliser la commande `plt.show()`

# Des conseils de bon usage
For non-interactive plotting it is suggested to use pyplot to create the figures and then the OO interface for plotting.
https://matplotlib.org/api/api_overview.html

L'utilisation de pyplot est plus simple (mais permet de faire des choses moins précises) que l'utilisation de matplotlib de façon orientée objet.

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linéaire')
plt.plot(x, x**2, label='carré')
plt.plot(x, x**3, label='cube')

plt.xlabel('abscisses')
plt.ylabel('ordonnées')

plt.title("Graphique basique")

plt.legend()

plt.show()
```

# Backend

`matplotlib` peut être utilisé dans beaucoup de situations différentes et nécessite des sorties différentes. (shell python avec des fenêtre pop up, Jupyter notebooks, graphiques embarqués dans des GUI comme wxpython ou pygtk , sorties d'images postscript).

Tous ces moyens sont des `backend`.

Le **frontend** est le code pour tracer le graphe. Le **backend** fait tout le travail "behind-the-scenes" pour produire la figure.

Il existe 2 types de backends: 
- **user interface backends** (pygtk, wxpython, tkinter, qt4, macosx) : ce sont les **interactive backends**.
- **hardcopy backends** pour créer des fichiers images(PNG, SVG, PDF, PS) : ce sont les **non-interactive backends**.

Pour savoir comment configurer son Backend (si nécessaire), consulter précisément la doc suivante : https://matplotlib.org/tutorials/introductory/usage.html#backends

Certains backends peuvent être activés dans un Jupyter notebook avec les fonctions magiques %matplolib 'nom du backend' (il y a un backend spécial Jupyter mais cela nécessite l'installation d'un module iympl https://github.com/matplotlib/jupyter-matplotlib).

Dans un environnement de notebook, il peut être intéressant de lister les backends existants pour trouver celui qui convient à notre besoin du moment (qui peut varier évidemment). Il faut redémarrer le kernel pour changer de backend. Voir la liste ci-dessous :

```python
% matplotlib -l
```

A priori, mes choix vont dans cet ordre :

1- notebook

2- inline

3- qt

Avec le backend notebook, il est conseillé de numéroter les figures pour éviter des soucis.

```python
%matplotlib notebook

x = np.linspace(0, 2, 100)

plt.figure(42)

plt.plot(x, x, label='linéaire')
plt.plot(x, x**2, label='carré')
plt.plot(x, x**3, label='cube')

plt.xlabel('abscisses')
plt.ylabel('ordonnées')

plt.title("Graphique basique")

plt.legend()

plt.show()
```

# Mode interactif

Il est possible (avec un backend interactif) de commuter le mode interactif ou non :
- plt.ion() : active le mode interactif : on peut modifier la figure "à chaud". La commande plt.draw() peut être nécessaire pour forcer le rafraîchissment de la figure.
- plt.ioff() : désactive le mode interactif : la figure n'apparaît que lorsque la commande plt.show() est exécutée. Cette commande bloque le script python en cours jusqu'à ce que la fenêtre graphique soit close.



Pour connaître l'état courant : plt.isinteractive()
