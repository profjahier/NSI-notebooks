y(x, t) = A * sin{2*Pi(t/T - x/lambda)}

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
Pi = np.pi
```

```python
lambda_ = 2 # longueur d'onde en m
x = np.linspace(0, 5*lambda_, 256)
```

```python
A = 1.5 # amplitude en ua
phi = 0 # d√©phasage (en multiple de Pi/4)
```

```python
y = A*np.cos(2*Pi*x/lambda_ + phi)
```

```python
plt.figure()
plt.plot(x, y, 'o-')
plt.ylabel('Amplitude (ua)')
plt.xlabel('Distance (m)')
plt.title('t = 0')
plt.show()
```

```python
%matplotlib inline
lambda_ = 2
x = np.linspace(0, 5*lambda_, 256)
phi = 0
for t in range(9):
    y = A*np.cos(2*Pi*x/lambda_ - phi)
    plt.figure()
    plt.plot(x, y)
    plt.scatter(t*lambda_/4, A, color='red')
    plt.scatter(t*lambda_/4+3*lambda_, A*math.cos(6*Pi), color='green')
    delta = 1.3
    plt.scatter(t*lambda_/4+delta*lambda_, A*math.cos(2*delta*Pi), color='black')
    plt.ylabel('Amplitude (ua)')
    plt.xlabel('Distance (m)')
    plt.title('t = '+str(t)+'T/4')
    plt.show()
    phi += Pi/2 # √† chaque 1/4 de p√©riode, le d√©phasage d'amplitude de chaque point du milieu est de Pi/2
```

```python
%matplotlib notebook
legende = ['0', 'T/4', 'T/2', '3T/4', 'T', '5T/4', '3T/2', '7T/4', '2T']
A = 1.5
lambda_ = 2
x = np.linspace(0, 5*lambda_, 256)
phi = 0
plt.figure()
for t in range(9):
    y = A*np.cos(2*Pi*x/lambda_ - phi)
    plt.subplot(3, 3, t+1)
    plt.plot(x, y)
    plt.scatter(t*lambda_/4, A, color='red')
    plt.scatter(t*lambda_/4+3*lambda_, A*math.cos(6*Pi), color='green')
    delta = 1.3
    plt.scatter(t*lambda_/4+delta*lambda_, A*math.cos(2*delta*Pi), color='black')
    plt.ylabel('Amplitude (ua)')
    plt.xlabel('Distance (m)')
    plt.title('t = '+legende[t])
    phi += Pi/2 # √† chaque 1/4 de p√©riode, le d√©phasage d'amplitude de chaque point du milieu est de Pi/2

plt.tight_layout() # ajoute de l'espace entre les Axes pour visualiser tout le texte (ticks et title)
plt.show()
```

```python
%matplotlib notebook
T = 2 # p√©riode en s
t = np.linspace(0, 5*T, 256)
phi = 0

plt.figure()
y = A*np.cos(2*Pi*t/T - phi)
plt.plot(t, y)
plt.ylabel('Amplitude (ua)')
plt.xlabel('Temps (s)')
plt.title('Amplitude en un point du milieu au cours du temps')

plt.show()
```
